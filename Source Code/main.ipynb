```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

df = pd.read_csv(r"C:\Users\shahj\OneDrive\Desktop\Programing\AI_ML (ChatGPT)\Projects\Student Success AI\Data\student_data.csv")
# print(df.head())


# -- For Missing Values -- 
# print(df.isnull().sum())  # Says : Give me total number of empty values in column wise

df.fillna(method="ffill", inplace=True)  # Says : Copy value from above, without creating new variable


# -- Encoding Data -- 
le = LabelEncoder()
df['internet_access'] = le.fit_transform(df['internet_access'])
df['parent_support'] = le.fit_transform(df['parent_support'])
df['health_issues'] = le.fit_transform(df['health_issues'])
df['result'] = le.fit_transform(df['result'])


# -- Feature Scaling --
# For : Scalling helps model to see all feature equally, no let big number bully small nums!
scaler = StandardScaler()
df[['study_hours', 'attendance']] = scaler.fit_transform(df[['study_hours', 'attendance']])


# -- Split Data --
x = df.drop('result', axis=1)
y = df['result']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)


# -- Train Model -- 
# For : Use LogisticRegression() when you need categorical data -> Yes/No
model = LogisticRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

# acc = accuracy_score(y_test, y_pred)
# cm = confusion_matrix(y_test, y_pred)
# report = classification_report(y_test, y_pred)

# print("Accuracy : ", acc)
# print("\nConfusion Matrix : ", cm)
# print("\nClassification Report : ", report)




# --- Decision Tree Classifier ---
from sklearn.tree import DecisionTreeClassifier

tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(x_train, y_train)

y_pred_tree = tree_model.predict(x_test)

acc_tree = accuracy_score(y_test, y_pred_tree)
cm_tree = confusion_matrix(y_test, y_pred_tree)
report_tree = classification_report(y_test, y_pred_tree)

print(" Decision Tree Accuracy:", acc_tree)
print("\n Confusion Matrix:\n", cm_tree)
print("\n Classification Report:\n", report_tree)


 
#  Save the Model

import joblib
import os

# Step: Create folder if not exists
os.makedirs("../models", exist_ok=True)

# Step: Save the trained model
joblib.dump(tree_model, "../models/student_model.pkl")

print("✅ Model saved successfully!")
```
